```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(['SibSp','Parch','Ticket'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Cabin'] = df['Cabin'].fillna('C100')
```

```python
df.columns
```

```python
df.head()
```

```python
df['Survived']=df['Survived'].map({
    0:'Died',
    1:'Survived'
})
```

```python
df.head()
```

```python
ax =sns.countplot(x='Pclass' ,hue ='Survived' ,data=df)
ax.set(title='Passenger status against passenger class',xlabel='class',ylabel='no of passengers')
plt.show()
```

```python
ax =sns.countplot(x='Sex' ,hue ='Survived' ,data=df)
ax.set(title='Surviavl rate of male vs female',xlabel='class',ylabel='no of passengers')
plt.show()
```

```python
interval=[0,18,35,50,120]
categories=['child','teen','adult','old']
df['age_cats']=pd.cut(df['Age'],interval,labels=categories)
ax =sns.countplot(x='age_cats' ,hue ='Survived' ,data=df)
ax.set(title='Passenger in each age group',xlabel='age categories',ylabel='no of passengers')
plt.show()
```
